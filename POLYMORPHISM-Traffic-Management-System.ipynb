```
# Smart Traffic Management System

# Parent Class
class TrafficDevice:
    def __init__(self, name):
        self.name = name

    def activate(self):
        print(self.name + " is activating...")


# Child Class 1
class TrafficLight(TrafficDevice):
    def __init__(self, name):
        super().__init__(name)

    def activate(self):
        print(self.name + ": Switching Red, Yellow, Green!")


# Child Class 2
class SpeedCamera(TrafficDevice):
    def __init__(self, name):
        super().__init__(name)

    def activate(self):
        print(self.name + ": Scanning and recording vehicle speed!")


# Child Class 3
class PedestrianSignal(TrafficDevice):
    def __init__(self, name):
        super().__init__(name)

    def activate(self):
        print(self.name + ": Walk signal is ON!")


# Extra Class
class EmergencySiren(TrafficDevice):
    def __init__(self, name):
        super().__init__(name)

    def activate(self):
        print(self.name + ": EMERGENCY! Clear the road!")


# Create one object of each class
light  = TrafficLight("Traffic Light")
camera = SpeedCamera("Speed Camera")
signal = PedestrianSignal("Pedestrian Signal")
siren  = EmergencySiren("Emergency Siren")

# Store in a list
devices = [light, camera, signal, siren]

# Activate all without checking types
print("--- Activating All Devices ---")
for device in devices:
    device.activate()
```
